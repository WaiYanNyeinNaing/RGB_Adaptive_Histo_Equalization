```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plt_imshow(title, image):
	# convert the image frame BGR to RGB color space and display it
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python
path = "images/street.jpg"
name = path.split("images/")[1]

image = cv2.imread(path)
(B, G, R) = cv2.split(image)

# show each channel individually
plt_imshow("Red", R)
plt_imshow("Green", G)
plt_imshow("Blue", B)
```

```python
# apply CLAHE (Contrast Limited Adaptive Histogram Equalization)
print("[INFO] applying CLAHE Red...")
clahe = cv2.createCLAHE(clipLimit=1,tileGridSize=(8,8))
equalized_red = clahe.apply(R)

print("[INFO] applying CLAHE Green...")
equalized_green = clahe.apply(G)

print("[INFO] applying CLAHE Blue...")
equalized_blue = clahe.apply(B)

# show each channel individually
plt_imshow("CLAHE Red", equalized_red)
plt_imshow("CLAHE Green", equalized_green)
plt_imshow("CLAHE Blue", equalized_blue)
```

```python
merged = cv2.merge([ equalized_blue,  equalized_green,  equalized_red])
plt_imshow("Merged", merged)
cv2.imwrite(f"result/histogram_equalized_{name}",merged)
```
